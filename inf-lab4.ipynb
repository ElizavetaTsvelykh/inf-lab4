```python
import numpy as np
matrix = np.random.randint(1, 20, size=(5, 7))
def process_matrix(matrix):
    result = matrix.copy()
    for r in range(matrix.shape[1]):  
        row = matrix[:, r]  
        max_element = row.max()  
        half_sum = (row[0] + row[-1]) / 2 
        result[max_element, r] = half_sum  
    return result
processed_matrix = process_matrix(matrix)
print(" :\n", matrix)
print(" :\n", processed_matrix)
```

```python
import numpy as np
matrix = np.random.randint(1, 20, size=(7, 5))
def process_matrix(matrix):
    result = matrix.copy()
    for col in range(matrix.shape[1]):  
        column = matrix[:, col] 
        max_element = column.max()  
        max_index = column.argmax()  
        half_sum = (column[0] + column[-1]) / 2  
        if max_element < half_sum:
            result[max_index, col] = half_sum  
        else:
            result[max_index, col] = max_index  
    return result
processed_matrix = process_matrix(matrix)
print(" :\n", matrix)
print(" :\n", processed_matrix)
```

```python
import numpy as np
n = 5
matrix = np.random.randint(1, 10, size=(n, n))
def extract_triangles(matrix):
    n = matrix.shape[0]
    upper_triangle = [] 
    lower_triangle = []
    for i in range(n):
        for j in range(n):
            if j >= i:  
                upper_triangle.append(matrix[i, j])
            if j <= i: 
                lower_triangle.append(matrix[i, j])
    return upper_triangle, lower_triangle

def print_triangle(matrix, triangle_type):
    n = matrix.shape[0]
    print(f"{triangle_type.capitalize()} triangle:")
    for i in range(n):
        if triangle_type == 'upper':
            print(" ".join(str(matrix[i, j]) if j >= i else " " for j in range(n)))
        elif triangle_type == 'lower':
            print(" ".join(str(matrix[i, j]) if j <= i else " " for j in range(n)))

upper_triangle, lower_triangle = extract_triangles(matrix)
print(" :\n", matrix)
print("\n   :")
print_triangle(matrix, 'upper')
print("\n   :")
print_triangle(matrix, 'lower')
```
